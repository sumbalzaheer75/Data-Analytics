## ***18. Which Workclass getting The Highest Salary***

```python
highest_salary_workclass = df.groupby('workclass')['income'].mean()
highest_salary_workclass_name = highest_salary_workclass.idxmax()
print(highest_salary_workclass_name)
highest_salary_value = highest_salary_workclass.max()
print(highest_salary_value)
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
gender_income = df.groupby('gender')['income'].mean()
male = gender_income['Male']*100
female = gender_income['Female']*100
print(f"Percentage of Males earning >50K: {male:.2f}%")
print(f"Percentage of Females earning >50K: {female:.2f}%")
if male > female:
    print('Males have better chance to get salary >50k')
elif female>male:
    print('Females have better chance to get salary >50k')
else:
    print('Both have equal chance to get salary >50k')
```
